```python
import pandas as pd
import numpy as np
import joblib
pd.options.display.float_format = '{:,.2f}'.format
import warnings
warnings.filterwarnings(action='ignore')
```

```python
with open('svc.pkl', 'rb') as file_1:
  svc = joblib.load(file_1)
  
with open('rf.pkl', 'rb') as file_2:
  rf = joblib.load(file_2)
  
with open('knn.pkl', 'rb') as file_3:
  knn = joblib.load(file_3)
  
with open('adb.pkl', 'rb') as file_4:
  adb = joblib.load(file_4)
```

['All Costs', 'Environmental Remediation Costs',
       'Public/Private Property Damage Costs', 'Net Loss (Barrels)',
       'Intentional Release (Barrels)', 'Unintentional Release (Barrels)',
       'Accident Latitude', 'Accident Longitude', 'Public Evacuations']

```python
np.random.seed(20) 
all_cost = np.random.randint(0,840526118,size=20)
environmental = np.random.randint(0,635000000,size=20)
public_cost = np.random.randint(0,23000000,size=20)
net_loss = np.random.randint(0,30565, size=20)
intentional_release = np.random.randint(0,70191, size=20)
unintentional_release = np.random.randint(0,30565,size=20)
latitude = np.random.uniform(18.45,70.26,size=20)
longitude = np.random.uniform(-158.10,104.26,size=20)
evacuations = np.random.randint(0,700, size=20)


data_inf = pd.DataFrame({'All Costs' : all_cost, 'Environmental Remediation Costs' : environmental, 
                        'Public/Private Property Damage Costs' : public_cost, 'Net Loss (Barrels)' : net_loss, 
                        'Intentional Release (Barrels)' : intentional_release, 'Unintentional Release (Barrels)' : unintentional_release, 
                        'Accident Latitude' : latitude, 'Accident Longitude' : longitude, 'Public Evacuations' : evacuations})
data_inf.head()
```

```python
data_inf['Cause SVC Predict'] = svc.predict(data_inf)
data_inf['Cause RandomForest Predict'] = rf.predict(data_inf)
data_inf['Cause KNN Predict'] = knn.predict(data_inf)
data_inf['Cause AdaBoost Predict'] = adb.predict(data_inf)
data_inf
```
